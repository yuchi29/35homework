```python
import numpy as np
import matplotlib.pyplot as plt

def quadratic_equation(a, b, c):
    x = np.linspace(-10, 10, 400)  
    y = a * x**2 + b * x + c  

    root1 = (-b + np.sqrt(b**2 - 4*a*c)) / (2*a)
    root2 = (-b - np.sqrt(b**2 - 4*a*c)) / (2*a)

    vertex_x = -b / (2*a)
    vertex_y = a * vertex_x**2 + b * vertex_x + c

    plt.plot(x, y, label=f'y = {a}x^2 + {b}x + {c}')
    plt.scatter([root1, root2], [0, 0], color='red', label='Roots')
    plt.scatter(vertex_x, vertex_y, color='green', label='Vertex')

    plt.annotate(f'Root 1: ({root1:.2f}, 0)', (root1, 0), xytext=(-70, 30), textcoords='offset points',
                 arrowprops=dict(arrowstyle='->', color='black'))
    plt.annotate(f'Root 2: ({root2:.2f}, 0)', (root2, 0), xytext=(-70, 30), textcoords='offset points',
                 arrowprops=dict(arrowstyle='->', color='black'))
    plt.annotate(f'Vertex: ({vertex_x:.2f}, {vertex_y:.2f})', (vertex_x, vertex_y),
                 xytext=(50, -40), textcoords='offset points',
                 arrowprops=dict(arrowstyle='->', color='black'))

    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Quadratic Equation')
    plt.legend()
    plt.grid(True)
    plt.show()

quadratic_equation(-3, 4, 7)
quadratic_equation(5, -4, 9)
```
